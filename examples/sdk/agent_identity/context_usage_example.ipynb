# Agent Identity SDK: Context Usage Example

This notebook demonstrates how the `AgentIdentityContext` operates and how you can manage state independently in different logical sections of your code.

```python
import asyncio
from agentarts.sdk import AgentArtsRuntimeContext
```

### Step 1: Base Context

When you set a value in the root scope, it acts as the baseline for the current execution thread.

```python
print("=== Main Context Setup ===")
AgentArtsRuntimeContext.set_user_id("user-main")
AgentArtsRuntimeContext.set_oauth2_callback_url("http://main.example.com/callback")
print(f"Current User ID: {AgentArtsRuntimeContext.get_user_id()}")
print(f"Current Callback URL: {AgentArtsRuntimeContext.get_oauth2_callback_url()}")
```

### Step 2: Coroutines & Isolation

Context variables act locally within async tasks, preserving state separation.

```python
async def isolated_task(task_name: str, new_user_id: str) -> None:
    print(f"\n[{task_name}] Started. Initial User ID inherited from parent: {AgentArtsRuntimeContext.get_user_id()}")

    # Set a new context specific to this task
    AgentArtsRuntimeContext.set_user_id(new_user_id)
    print(f"[{task_name}] Updated User ID: {AgentArtsRuntimeContext.get_user_id()}")

    # Simulate some work
    await asyncio.sleep(0.5)
    print(f"[{task_name}] Finished. Final User ID: {AgentArtsRuntimeContext.get_user_id()}")


print("\n=== Spawning Isolated Tasks ===")
await asyncio.gather(
    isolated_task("Task A", "user-task-a"),
    isolated_task("Task B", "user-task-b")
)

print("\n=== Back to Main Context ===")
print(f"After tasks, Main User ID is still untouched: {AgentArtsRuntimeContext.get_user_id()}")
```
